```python
!pip install transformers
!pip install pandas
import pandas as pd

# Read the Excel file into a DataFrame
df = pd.read_excel('EXAMPPLE EXCEL SHEET ')
df = df.applymap(str)
# Extract the text samples from the DataFrame
text_samples = df['Text'].tolist()

from transformers import pipeline

# Load the classification pipeline
generator = pipeline(model="cardiffnlp/twitter-roberta-base-sentiment")

# Use the pipeline to classify the text samples
truncation=True
predictions = generator(text_samples, max_length=512)

# Create a new DataFrame with the classification results
results_df = pd.DataFrame(predictions)

# Save the DataFrame to an Excel file
results_df.to_excel('classification_results(EXCEL SHEET NAME).xlsx', index=False)
```
